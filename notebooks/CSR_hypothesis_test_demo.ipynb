# CSR Hypothesis Test Demo

Quick demo to show how to use the `CSR_hypothesis_test_dmin` function. We use the average minimum neighbour distance statistic:
$$\bar{d}_{\text{min}}:= \dfrac{1}{N}\sum_{i=1}^{N}d_{\text{min}}(u_i)$$

## Testing with a Homogeneous Poisson Process

Generate a homogeneous Poisson process. We expect this to conform well to the CSR hypothesis

```python
observed_pp = hom_poisspp(100, window={'x_min':0.0, 'x_max':1.0, 'y_min':0.0, 'y_max':1.0})
```

```python
observed_pp.num_points
```

Call the function! Here we use a significance level of 0.05 and simulate the process 1000 times. For convenience, the plot of the distribution is also returned.

```python
CSR_hypothesis_test_dmin(observed_pp, 0.05, n_sim=1000, plot_dist=True)
```
