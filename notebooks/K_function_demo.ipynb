# K Function Demo (Radial Neighbours Function)

```python
from point_process_utils import Point_Process, Point, hom_poisspp
from summary_statistics import K_hat
import seaborn as sbn
import matplotlib.pyplot as plt
import numpy as np
```

Create rectangular window.

```python
window = {'x_min': 0.0, 'x_max': 10.0, 'y_min':0.0, 'y_max': 10.0}
```

Generate a homogeneous Poisson Process, with intensity 0.05. Check to see how many points have been generated.

```python
pp = hom_poisspp(1, window)                                               
pp.num_points
```

Choose domain to calculate $K$ function on

```python
r = np.linspace(0, 3, 100)
```

Call the function with the option to plot the result. Points returned in array.

First we do not restrict the domain.

```python
K_hat_array = K_hat(pp, r, restrict_domain=False, plot=True)
```

Now, restrict the domain so that edge effects do not occur.

```python
K_hat_array_2 = K_hat(pp, r, restrict_domain=True, plot=True)
```
