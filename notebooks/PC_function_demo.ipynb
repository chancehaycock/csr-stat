# Pair Correlation Function Demo

```python
from point_process_utils import Point_Process, Point, hom_poisspp
from summary_statistics import O_hat, PC_hat
import seaborn as sbn
import matplotlib.pyplot as plt
import numpy as np
```

Create rectangular window.

```python
window = {'x_min': 0.0, 'x_max': 10.0, 'y_min':0.0, 'y_max': 10.0}
```

Generate a homogeneous Poisson Process, with intensity 3. Check to see how many points have been generated.

```python
pp = hom_poisspp(3, window)                                               
pp.num_points
pp.display()
```

Choose domain to calculate $PC$ on, same as domain to calculate $O$ on

```python
r = np.linspace(0, 3, 100)
```

Call the function with the option to plot the result. Points returned in array.

First we look at the default standard box kernel

```python
PC_array = PC_hat(pp, r, bandwidth=0.2, plot=True)
```

Now, lets check out Epanechnikov kernel

```python
PC_array2 = PC_hat(pp, r, bandwidth=0.2, plot=True, kernel="EK")
```

results seem a bit smoother. Lets try both without resticting domain

```python
PC_array3 = PC_hat(pp, r, bandwidth=0.2, plot=True, restrict_domain=False)
```

```python
PC_array4 = PC_hat(pp, r, bandwidth=0.2, plot=True, restrict_domain=False, kernel="EK")
```
