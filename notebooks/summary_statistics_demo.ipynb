# Summary Statistics Envelope Simulation Demo

this notebook demonstrates the use and performance of simulate_summary_statistic, which compares the plot of a metric for a dataset to an n_sims average of a simualted homogenous poisson point process in the same window with the same (estimated) global intensity

```python
import matplotlib.pyplot as plt
import numpy as np
import scipy as sp
import seaborn as sbn
from point_process_utils import *
from summary_statistics import *
```

 create window & homogenous poisson point process

```python
window = {'x_min': 0.0, 'x_max': 10.0, 'y_min':0.0, 'y_max': 10.0}        
pp = hom_poisspp(2, window)     
r = np.linspace(0, 3, 100)
```

Run our simulate_summary statistic for F, G, K & L, plotting one for n_sims with a min/max envelope 

```python
f=simulate_summary_statistic(pp, r, F_hat, n_sims=40, plot=True)
```

F(r) appears to work well, given example is inside envelope

```python
print(f)
```

summary statistic above demonstrates the tuple returned, containing the mean difference and max difference

```python
f=simulate_summary_statistic(pp, r, G_hat, n_sims=100, plot=True)
```

G(r) also appears to be working well, given example is inside envelope

```python
print(f)
```

```python
f=simulate_summary_statistic(pp, r, K_hat, n_sims=40, plot=True)
```

K(r) appears to work well, given example is inside envelope. Running time for K stat is higher than F & G (why?). The max/min envelope increases with r, which means the uncertainty of K(r) decreases with r (becomes a less useful measure?)

```python
print(f)
```

```python
f=simulate_summary_statistic(pp, r, L_hat, n_sims=40, plot=True)
```

L(r) appears to work well, given example is inside envelope. Running time for L is also high, no suprise due to the fact it is derived from K. The max/min envelope also iincreases with r. Slight downwards trend with r might sugggest that the edge effect isn't fully counted for, but this should probably be probed with a higher n_si
